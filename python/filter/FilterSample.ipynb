# フィルタ処理サンプル

乱数波形を用いて，LowPassFilter, HighPassFilter, BandPassFilter, BandStopFilterの挙動を確認する．

```python
import matplotlib.pyplot as plt
from lib.filter import signal_filter, signal_fft
from lib.create_signal import *
```

## フィルタ動作確認用サンプル波形

```python
FS = 48000
DURATION = 1
t, data = create_signal(FS, DURATION)
```

```python
plt.figure()
plt.plot(t, data)
plt.xlabel('time [sec]')
plt.ylabel('amplitude')
plt.show()
plt.close()
```

```python
freq, data_fft = signal_fft(t, data)
```

```python
plt.figure()
plt.plot(freq, data_fft.real)
plt.xlabel('freq [Hz]')
plt.ylabel('spectrum')
plt.show()
plt.close()
```

## LowPassFilter

```python
class LowPassFilterParameters():
    freq_pass = 3000
    freq_stop = 6000
    gain_pass = 3
    gain_stop = 40
    btype = 'lowpass'
```

```python
param = LowPassFilterParameters()
data_lpf = signal_filter(data, FS, param.freq_pass, param.freq_stop, param.gain_pass, param.gain_stop, param.btype)
```

```python
plt.figure()
plt.plot(t, data_lpf)
plt.xlabel('time[sec]')
plt.ylabel('amplitude')
plt.show()
plt.close()
```

```python
freq, data_lpf_fft = signal_fft(t, data_lpf)
```

```python
plt.figure()
plt.plot(freq, data_lpf_fft.real)
plt.xlabel('freq [Hz]')
plt.ylabel('spectrum')
plt.show()
plt.close()
```
