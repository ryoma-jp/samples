## データセットの読み込み(ボストン住宅価格)

````python
from sklearn.datasets import load_boston
import pandas as pd
from sklearn.model_selection import train_test_split

boston = load_boston()
df_x_boston = pd.DataFrame(boston['data'], columns=boston['feature_names'])
df_y_boston = pd.DataFrame(boston['target'], columns=['MEDV'])
df_boston = pd.concat([df_x_boston, df_y_boston], axis=1)

x_train, x_test, y_train, y_test = train_test_split(df_x_boston, df_y_boston, test_size=0.3, random_state=0)

df_train = pd.concat([x_train, y_train], axis=1)
df_test = pd.concat([x_test, y_test], axis=1)
````

## AutoKeras動作確認

````python
import autokeras as ak
````

````python
# Initialize the structured data regressor.
reg = ak.StructuredDataRegressor(
    overwrite=True, max_trials=100,
    project_name="structured_data_regressor",
    metrics=["mean_squared_error", "mean_absolute_error"],
    seed=1234
)  # It tries 10 different models.
# Feed the structured data regressor with training data.
reg.fit(
    x=x_train,
    y=y_train,
    epochs=200,
)
````

````python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

pred = reg.predict(x_test)
r2 = r2_score(y_test, pred)
mse = mean_squared_error(y_test, pred)

print("R2 : %.3f" % r2)
print("MSE : %.3f" % mse)
````

次にデータを標準化して試してみる．

````python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)
x_train_std = pd.DataFrame(sc.transform(x_train), columns=boston['feature_names'])
x_test_std = pd.DataFrame(sc.transform(x_test), columns=boston['feature_names'])
````

ついでにTensorBoardで可視化もできるみたいなので，試してみる．

````python
std_dir='structured_data_regressor-std'
!mkdir -p $std_dir
!ls
````

````python
from tensorflow import keras

tensorboard_callback_train = keras.callbacks.TensorBoard(log_dir=std_dir)
````

````python
# Initialize the structured data regressor.
reg_std = ak.StructuredDataRegressor(
    overwrite=True, max_trials=100,
    project_name=std_dir,
    metrics=["mean_squared_error", "mean_absolute_error"],
    seed=1234
)  # It tries 10 different models.
# Feed the structured data regressor with training data.
reg_std.fit(
    x=x_train_std,
    y=y_train,
    epochs=200,
    callbacks=[tensorboard_callback_train],
)
````

````python
pred_std = reg_std.predict(x_test_std)
r2_std = r2_score(y_test, pred_std)
mse_std = mean_squared_error(y_test, pred_std)

print("R2 : %.3f" % r2_std)
print("MSE : %.3f" % mse_std)
````
